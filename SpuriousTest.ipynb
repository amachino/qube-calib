```python
import bokeh.io
import bokeh.plotting
import bokeh.models
from bokeh.palettes import Spectral8, Category20
import pickle

bokeh.plotting.output_notebook()

offset = lambda l, d: [v + i*d for i, v in enumerate(l)]

def plot_spurious_multi(p, d):
    p.renderers = []
    p.multi_line(offset(d['freq'],0), offset(d['trace'],0), line_color=Spectral8)
    bokeh.plotting.show(p)
    
def plot_spurious(p, d, i, c):
    p.renderers = []
    p.line(d['freq'][i], d['trace'][i], line_color=c)
    bokeh.plotting.show(p)

def plot_spurious_lo(f, v, p):
    p.multi_line(f, v, line_color=Spectral8)
    bokeh.plotting.show(p)
    
## 測定データの復旧

for FL in range(10400, 11001, 200):
    freq,powr = [],[]
    print('LO: {}'.format(FL))
    for FI in range(1800,3201,200):
        FR = FL - FI
        with open('if{}rf{}.pickle'.format(FI, FR), 'rb') as f:
            d = pickle.load(f)
        #print('LO: {}, IF: {}, RF: {}'.format(FL,FI,FR))
        #print(FL + FI, FL - FI, 2*FL+FI, 2*FL-FI, FL+2*FI, FL-2*FI, )
        freq.append(d['freq'][1])
        powr.append(d['trace'][1])
    p = bokeh.plotting.figure(y_range=[-80, 10], height=100, toolbar_location=None, sizing_mode="stretch_width")
    plot_spurious_lo(freq, powr, p)
    #plot_spurious(p, d, 1)
```
