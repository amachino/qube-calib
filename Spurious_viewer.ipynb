```python
import numpy
import matplotlib.pyplot as plt
import pickle
%matplotlib inline

def plot_spurious_multi(d, ox=10, oy=0):
    plt.figure(figsize=(12,4))
    for i, v in enumerate(d['vatt']):
        plt.plot(d['freq'][i] + i*ox, d['trace'][i] + i*oy, linewidth = 0.5
                 , color='blue')
    return d['vatt']
    
def plot_spurious(d, i):
    plt.plot(d['freq'][i], d['trace'][i])
    return d['vatt'][i]
```

```python
with open('if2000rf8600.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
plot_spurious_multi(data)
```

```python
plot_spurious(data, 8)
```
